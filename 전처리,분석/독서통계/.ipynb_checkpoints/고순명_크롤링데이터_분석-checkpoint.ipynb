```python
from surprise.model_selection import GridSearchCV
from surprise.model_selection import train_test_split
import pandas as pd
from surprise import Dataset, Reader, SVD
import matplotlib.pyplot as plt
import seaborn as sns
# 한글 폰트
import matplotlib as mpl
mpl.rc('font', family='Malgun Gothic')
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('크롤링_원본_합본_책소개포함.csv')
```

```python
top_genres = df["genre"].value_counts().index

# 해당 장르들의 평균 평점 계산
genre_avg_rating = df[df["genre"].isin(top_genres)].groupby('genre')['rating'].mean().sort_values(ascending=False)

# 시각화
plt.figure(figsize=(12, 8))
sns.set_palette(palette="pastel")
sns.barplot(x = genre_avg_rating.values, y = genre_avg_rating.index)
plt.xlabel("평균 평점")
plt.ylabel("장르")
plt.title("장르별 평균 평점")
plt.grid(axis = 'x', linestyle = '--', alpha = 1)
plt.tight_layout()
plt.show()
```

```python
genre_counts = df['genre'].value_counts().reset_index()
genre_counts.columns = ['genre', 'count']

genre_ratings = df.groupby('genre')['rating'].mean().reset_index()
genre_ratings.columns = ['genre', 'avg_rating']

genre_summary = pd.merge(genre_counts, genre_ratings, on='genre')

genre = genre_summary.sort_values(by='count', ascending=False)

plt.figure(figsize=(14, 8))
sns.set_palette(palette="pastel")
sns.barplot(data=genre, x='count', y='genre')
plt.title('장르별 도서 수')
plt.xlabel('도서 수')
plt.ylabel('장르')
plt.grid(axis='x', linestyle='--', alpha=1)
plt.tight_layout()
plt.show()
```

```python
df_rating = df.groupby('rating').size().reset_index(name='count')
df_rating = df_rating.set_index('rating')

df_rating.plot(kind='bar')
plt.title('평점 분포')
plt.xlabel('평점')
plt.ylabel('빈도')
```

```python
df['user_mean'] = df.groupby('user_id')['rating'].transform('mean')

df['rating_cen'] = df['rating'] - df['user_mean']

user_bias = df.groupby('user_id')['rating_cen'].mean().reset_index()
user_bias.columns = ['user_id', 'mean_centered_rating']


plt.figure(figsize=(10, 6))
sns.histplot(user_bias['mean_centered_rating'], bins=30, kde=True)
plt.title('유저별 중심화 평점 평균 분포')
plt.xlabel('중심화 평점 평균')
plt.ylabel('유저 수')
plt.show()
```

```python
# 유저들의 평점이 대부분 일관된다
```

```python
top_good = user_bias.sort_values(by='mean_centered_rating', ascending=False).head(10)
top_bad = user_bias.sort_values(by='mean_centered_rating').head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x='user_id', y='mean_centered_rating', data=top_good)
plt.xticks(rotation=45)
plt.title('평점에 후한 유저 Top 10')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x='user_id', y='mean_centered_rating', data=top_bad)
plt.xticks(rotation=45)
plt.title('평점에 박한 유저 Top 10')
plt.show()
```
